```python
import pandas_datareader.data as web
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = web.DataReader('AAPL', 'yahoo', start='2020-01-13', end='2022-01-13')['Close']
df = {'Close Price': df}
df = pd.DataFrame(df)
df.head(10)
```

```python
# Bull markets is a market condition where investors are expecting prices to rise.
# Bear markets is a market condition where investors are expecting prices to fall.
# 
# Calculating SMA (simple moving average *apply equal weight to all data points)
#
# Close price is different from adjusted close price because adjusted close price takes into consideration dividends paid.
#
# Closing Price is the last price at which a stock trades during a regular trading session
```

```python
# We need to get 100 data points from the date provided. 
```

```python
# end = dt.datetime.today()
```

```python
df.plot(figsize = (10, 10))
plt.grid()
plt.xlabel("Date")
plt.ylabel("Price in USD")
plt.show()
```

```python
df['50_SMA'] = df['Close Price'].rolling(window = 30, min_periods = 1).mean()
df['100_SMA'] = df['Close Price'].rolling(window = 50, min_periods = 1).mean()
df
```
